```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import model_selection as ms
pd.set_option("display.max_rows",15)
%matplotlib inline
```

```python
train_size = 84195
sample_data = pd.read_csv("D:/AI/merged_5s_A.csv", header = 0, nrows = train_size)
sample_data.to_pickle('D:/AI/merged_5s_A.pkl')
```

```python
train = pd.read_pickle('D:/AI/merged_5s_A.pkl')
train_label = train.label.copy()

train
```

```python
train_label.unique()
```

```python
category_variables = ["label"]

for cv in category_variables:
    train[cv] = train[cv].astype("category")
    #dataset.kdd_test[cv] = dataset.kdd_test[cv].astype("category", categories = dataset.kdd_train[cv].cat.categories)
    
print("Length of Categories for {} are {}".format(cv , len(train[cv].cat.categories)))
print("Categories for {} are {} \n".format(cv ,train[cv].cat.categories))
```

```python
dummy_variables_2labels = category_variables
    
class preprocessing:
    train_labels = pd.get_dummies(train, columns = dummy_variables_2labels, prefix=dummy_variables_2labels)
    #kdd_test_2labels = pd.get_dummies(dataset.kdd_test, columns = dummy_variables_2labels, prefix=dummy_variables_2labels)
```

```python
preprocessing.train_labels.to_pickle("D:/AI/preprocessed_train_data_A.pkl")
#preprocessing.kdd_test_2labels.to_pickle("dataset/preprocessed_test_data.pkl")

preprocessing.train_labels
```
