```python
category_variables = ["Protocol"]

for cv in category_variables:
    b4[cv] = b4[cv].astype("category")
    #a5[cv] = a5[cv].astype("category")
    
    print("Length of Categories for {} are {}".format(cv , len(b4[cv].cat.categories)))
    print("Categories for {} are {} \n".format(cv ,b4[cv].cat.categories))
```
